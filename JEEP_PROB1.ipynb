### PROBLEM:1
  Solving the jeep problem,for a  fixed fuel availabilty and then maximizing the distance travelled into the desert
<br>

```python
   #TANK CAPACITY IS FIXED-MAXIMUM DISTANCE THAT CAN BE TRAVELLED BY VARYING NO OF DRUMS AVAILABLE AT START(N)
                      #THIS CAN BE FORMULATED AS A LINEAR PROGRAMMING PROBLEM
    
    
#variables required
# n=no on full drums given at start 
# c=capacity of the jeep(here assumed to be 5)
# dist[i]= distance between i and i+1 th point


from pulp import *

M=[4,5,6,8]
for n in M:
    # creating a problem object/decision variable in pulp
    problem=LpProblem("jeep",LpMaximize)

    #creating variables of the problem
    #SOLVING FOR DIFFERENT VALUES OF N(NO OF DRUMS CARRIED BY JEEP)
    dist=LpVariable.dicts("d", list(range(n)), 0,min(5+1,5*n))


    #creating constraints for the problem
    #constraint :0
    sum1=0
    for i in range(n):
        sum1+=(2*n-(2*i+1))*dist[i]
    c1=sum1<=5


    #constraints :[k,j]
    C=[]
    for k in range(1,n):
        for j in range(k+1,n+1):
            sum=0
            for l in range(k+1,n+1):
                if(l!=(j+1)):
                    sum+=(2*(n-j+1)-1)*dist[l-1]
                else:
                    sum+=(2*(n-j-1)-1)*dist[l-1]
            C.append(sum<=min(5*(n-k)+(n-j+1),5*n))



     # alternative for of above constraints(in recursive form):k=0,1.....n-1
    Q=[]
    for k in range(n):
        lhs=0
        rhs=(5*(n-k)+1)
        for j in range(k+1,n+1):
            lhs += (2*(n-k)-1)*dist[j-1]
            sum=0
            for l in range(k+1,j+1):
                sum+=dist[l-1]
            if(2*sum<=1):
                rhs += 2*sum
            else:
                rhs +=1
        Q.append(lhs<=rhs)





    #adding constraints to the problem
    problem += c1
    for i in range(len(Q)):
        problem += Q[i]

    #defining the objective function
    objective=0
    for i in range(n):
        objective +=dist[i]
    problem+=objective

    print(problem)

    problem.solve()

    #for i in range(n):
    # print(f"dist {i}: {dist[i].varValue}","units")

    sum2=0
    for i in range(n):
        sum2+=dist[i].varValue

    print("maximum distance that can be travelled by carrying","n =",n,"drums is",sum2+13*(n-4.01)/5,"units")
    
    
#note:the problem can be generalised for varying tank capacity c by replacing 5 with variable c
```
